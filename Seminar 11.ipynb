```python
#1_1
import matplotlib.pyplot as plt
   
Country = ['Python','Java','JavaScript','C#','C/C#','PHB','R','Other']
GDP_Per_Capita = [12.9,10.12,2.3,6.4,19.53,1.5,1.58,43.91]

New_Colors = ['green','blue','purple','brown','teal','pink','yellow','grey']
plt.bar(Country, GDP_Per_Capita,color=New_Colors)
plt.title('Top computer languages-TIOBE (Dec 2021)')
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
#1_2
import matplotlib.pyplot as plt
import numpy as np

y = np.array([12.9,10.12,2.3,6.4,19.53,1.5,1.58,43.91])
mylabels = ['Python','Java','JavaScript','C#','C/C#','PHB','R','Other']

plt.pie(y, labels = mylabels, startangle=90, autopct='%1.1f%%')
plt.title('Top computer languages-TIOBE (Dec 2021)')
plt.show() 
```
