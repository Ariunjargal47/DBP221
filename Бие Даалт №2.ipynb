```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
dt = pd.read_excel('REPORT 2.xlsx', sheet_name = 6)
data1 = dt.head(-1)
avg = dt['Нийт'].tail(-28)
data2 = pd.read_excel('REPORT 2.xlsx', sheet_name = 7)
```

```python
#1 Оюутнуудын курс
student = data2['Оюутны код']
sid = student.values.tolist()
listid = []
for i in sid:
    a = int(i[1:3])
    if a == 17:
        listid.append('5-р курс')
    if a == 16:
        listid.append('6-р курс')
    if a == 18:
        listid.append('4-р курс')
    elif a == 19:
        listid.append('3-р курс')
data2['Курс'] = listid
data2
```

```python
#2 Шалгалт ба Ирц регресс
x1 = pd.DataFrame(data1['Ирц-10'])
y1 = pd.DataFrame(data1['Шалгалт-30'])
mod1 = sm.ols('y1 ~ x1', data1).fit()
print(mod1.summary())
```

```python
#3 Судалсан хичээлийн тоо болон голч дүнгийн регресс
x2 = pd.DataFrame(data2['Судалсан хичээлийн тоо'])
y2 = pd.DataFrame(data2['Голч дүн'])
mod2 = sm.ols('y2 ~ x2', data2).fit()
print(mod2.summary())
```

```python
#4 Ерөнхий хамгийн бага квадратын аргын үнэлгээ
x3 = pd.DataFrame(data2['ЭЕШ оноо'])
mod3 = sm.ols('y2 ~ x3', data2).fit()
print(mod3.summary())
print(mod3.params)
x4 = list(data2['ЭЕШ оноо'])
y3 = list(data2['Голч дүн'])
plt.scatter(x3, y2, color = 'darkblue')
plt.xlabel('ЭЕШ оноо')
plt.ylabel('Голч дүн')
z = np.polyfit(x4, y3, 1)
p = np.poly1d(z)
plt.plot(x3,p(x3),"--g")
plt.show()
```

```python
#5 Голч дүнгийн таамаглал
df0 = pd.DataFrame(mod3.params)
df1 = df0.transpose()
reg = df1['Intercept'] + df1['x3'] * 620
reg
```

```python
#6 Barchart үүсгэх нь
plt.figure(figsize = (8,4))
plt.bar('Оюутны нэр', 'Нийт', data = data1, color = 'blue')
plt.xticks(rotation = 90)
#7 Linechart-Дундаж
plt.hlines(avg, 0 ,28, color = 'red', label = 'Дундаж')
plt.legend()
plt.xlabel('Оюутны нэр')
plt.ylabel('Дүн')
```

```python
#8 Үсгэн үнэлгээний давтамж
ratings = data1.iloc[0:5, 12:14].set_axis(['Давтамж', 'Үнэлгээ'], axis = 1)
plt.pie('Давтамж', data = ratings, labels = 'Үнэлгээ', autopct = "%.f")
plt.title('Үнэлгээ')
```

```python
#9 Мэргэжил тус бүр дээрх оюутнуудын тоо
year = pd.DataFrame(data2['Суралцаж буй мэргэжил'].value_counts()).reset_index().rename(columns = {'index' : 'Суралцаж буй мэргэжил', 'Суралцаж буй мэргэжил': 'Давтамж'})
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (25,5))
ax1.pie('Давтамж', labels = 'Суралцаж буй мэргэжил', data = year, autopct = "%.f")
ax1.set_title('Суралцаж буй мэргэжил')
ax2.bar('Суралцаж буй мэргэжил', 'Давтамж', data = year)
ax2.set_xlabel('Мэргэжил')
ax2.set_ylabel('Давтамж')

```
